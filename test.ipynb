```python
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

# Dữ liệu và nhãn mẫu
data = ["This is spam", "Hello friend!", "Buy now!", "Win a free car!", "Good morning"]
labels = [1, 0, 1, 1, 0]  # 1: spam, 0: ham

# Xử lý dữ liệu
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data)

# Huấn luyện mô hình
model = LogisticRegression()
model.fit(X, labels)

# Lưu mô hình Logistic Regression vào file .pkl
with open('logistic_model.pkl', 'wb') as file:
    pickle.dump(model, file)

# Lưu CountVectorizer vào file .pkl
with open('vectorizer.pkl', 'wb') as file:
    pickle.dump(vectorizer, file)

print("Mô hình và vectorizer đã được lưu thành công!")
```
